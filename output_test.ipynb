```python
from IPython.display import HTML

tree_data = """
const treeData = {
  "name": "JPW9602804 / WO1996JP02804",
  "children": [
    {
      "name": "EP1996932019",
      "type": "default",
      "children": [
        { "name": "EP1999104107", "type": "default",
          "children": [
            { "name": "EP2004008669", "type": "default",
              "children": [
                { "name": "EP2009177103", "type": "default" }
              ]
            }
          ]
        }
      ]
    },
    {
      "name": "JPW9700615 / WO1997JP00615",
      "type": "priority",
      "children": [
        { "name": "WO1997JP04429", "type": "default" },
        { "name": "EP1997903634", "type": "default" }
      ]
    }
  ]
};
"""

html_code = f"""
<div id="tree"></div>
<script src="https://d3js.org/d3.v7.min.js"></script>
<script>
{tree_data}

// SVG setup
const width = 800, height = 600;
const svg = d3.select("#tree").append("svg")
    .attr("width", width)
    .attr("height", height)
    .append("g")
    .attr("transform", "translate(50,50)");

const treeLayout = d3.tree().size([width - 100, height - 100]);
const root = d3.hierarchy(treeData);
treeLayout(root);

// Links
svg.selectAll("line")
    .data(root.links())
    .enter()
    .append("line")
    .attr("x1", d => d.source.x)
    .attr("y1", d => d.source.y)
    .attr("x2", d => d.target.x)
    .attr("y2", d => d.target.y)
    .attr("stroke", "#ccc");

// Nodes
svg.selectAll("circle")
    .data(root.descendants())
    .enter()
    .append("circle")
    .attr("cx", d => d.x)
    .attr("cy", d => d.y)
    .attr("r", 5)
    .attr("fill", d => d.data.type === "priority" ? "orange" : "blue")
    .on("mouseover", function(event, d) {{
        d3.select(this).attr("r", 8);
    }})
    .on("mouseout", function(event, d) {{
        d3.select(this).attr("r", 5);
    }});

// Labels
svg.selectAll("text")
    .data(root.descendants())
    .enter()
    .append("text")
    .attr("x", d => d.x + 10)
    .attr("y", d => d.y + 5)
    .text(d => d.data.name)
    .style("font-size", "12px");
</script>
"""

HTML(html_code)
```
